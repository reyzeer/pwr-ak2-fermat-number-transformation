```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({
    'image.cmap': 'gray',
    'figure.dpi': 120,
    'axes.grid': False,
})
```

```python
from skimage import data
from skimage.color import rgb2gray

original_img = data.imread('glowa.jpg')
img = rgb2gray(original_img)

fig, axs = plt.subplots(1,2,figsize = (15,8))
axs[0].imshow(original_img)
axs[1].imshow(img)
```

```python
loc = (500, 400)
radius = 300

s = np.linspace(0, 2*np.pi, 1000)
x = loc[0] + radius * np.cos(s)
y = loc[1] + radius * np.sin(s)
snakes = np.array([x,y]).T

fig, axs = plt.subplots(1,4, figsize = (12,3))
axs[0].plot(s)
axs[1].plot(x)
axs[2].plot(y)
axs[3].plot(snakes)
```

```python
from skimage.filters import gaussian
blurred_image = gaussian (img, 10)

plt.imshow(blurred_image)
plt.plot(snakes[:,0],snakes[:,1])
```

```python
from skimage.segmentation import active_contour
fig, axs = plt.subplots(1,2, figsize = (14,7))
contour = active_contour(blurred_image, snakes)
                        
axs[0].plot(snakes)
axs[1].plot(contour)
```
